```python
# nd_array_ 
# support GPU
# immutable

import tensorflow as tf
```

```python
# 1 -   create tensors
 # scalar, rank 0
x = tf.constant(4)
print(x)
```

```python
x = tf.constant(4, shape = (1,1), dtype=float)
print(x)
```

```python
# vector, rank 2 
x = tf.constant([1,2,3])
print(x)
```

```python
# matrix, rank 2
x = tf.constant([[1,2,3], [4,5,3]])
print(x)
```

```python
tf.random.normal((3,3), mean=0, stddev=1)
print(x)
```

```python
x = tf.range(10)
print(x)
```

```python
x = tf.constant([1,2,3])
y = tf.constant([4,5,6])
z = x + y
print(z)
z = x/y
print(z)
z = tf.tensordot([1,2],[0,1], axes=1)
print(z)
```

```python
# indexing
x = tf.constant([[1,7,3,4],[1,2,8,5]])
print(x[0][2])
print(x[:,0]) # first element of each column
print(x[1, 1:3]) 
print(x[0,1])
print(x[1,:])
```
